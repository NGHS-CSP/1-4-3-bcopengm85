Whether the array for each pixel has three or four elements, a color image is a 3-dimensional array:

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import PIL
import os.path

directory = os.getcwd()
filename = os.path.join(directory, 'woman.jpg')
# Read the image data into an array
img = plt.imread(filename)

def make_mask(rows, columns, stripe_width):
    '''An example mask generator
    Makes slanted stripes of width stripe_width
    image
    returns an ndarray of an RGBA image rows by columns
    '''
    
    img = PIL.Image.new('RGBA', (columns, rows))
    image = np.array(img)
    for row in range(rows):
        for column in range(columns):
            if column in range(60, 100) or row in range(65,95):
                image[row][column] = [0, 0, 225, 255]
            elif column in range(50, 60) or column in range(100, 110) or row in range(55, 65) or row in range(95, 105):
                image[row][column] = [255, 255, 255, 0]
            else:
                image[row][column] = [250, 0, 25, 255]
    
    return image


image = make_mask(160,270,15)
    
fig, ax = plt.subplots(1, 1)

make_mask(100, 200, 30)
ax.imshow(image, interpolation='none')
````
